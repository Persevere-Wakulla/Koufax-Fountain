Funtion and a table of (x,y) values

```python
x = 5
y = 4*x + 3 # The function

print(x,",",y)

print("x \t y")

for x in range(11):
    y = 4*x + 3 # The fucntion
    print(x, "\t", y)  
```

Defining a Python function

```python
def f(x):
    y = 4*x + 3
    return y

print(5, ",", f(5))

# Next, a loop
for x in range(11):
    print(x, "\t", f(x)) 
```

Display axis lines

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax], [0,0],'b') # blue x axis
plt.plot([0,0], [ymin,ymax],'b') # blue y axis

plt.show()
```

Plot several points as a function

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax], [0,0],'b') # blue x axis
plt.plot([0,0], [ymin,ymax],'b') # blue y axis

for x in range(10):
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')
    
plt.show()
```

Graph and table of (x,y) values

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax], [0,0],'b') # blue x axis
plt.plot([0,0], [ymin,ymax],'b') # blue y axis

print("x \t y")
for x in range(xmin, xmax+1):
    y = 0.5*x + 1
    plt.plot([x],[y], 'ro')
    print(x, "\t" ,y)
    
plt.show()
```
