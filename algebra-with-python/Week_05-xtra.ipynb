25. A towns population increases at a constant rate. In 2010 the population was 55,000. By 2012 the population had increased to 76,000. If this trend continues, predict the population in 2016.

26. The number of people afflicted with the commo cold in the winter months dropped steadily by 50 each year since 2004 until 2010, in 2004, 875 people were inflicted.

Find the linear function that models the number of people afflicted with the common cold, C, as a function of the year, t. When will no one be afflicted?



Graph the solution - Review exercise #25

```python
import matplotlib.pyplot as plt
import numpy as np

# Only thing needed to change is x & y points⬇️
x1 = 0
y1 = 55
x2 = 2
y2 = 76
# Only thing needed to change is x & y points⬆️

# Develop the equation y = mx +b
m = (y2 - y1) / (x2 -x1)
b = y1 - m*x1
print("y = ", m, "x + ", b)

# For the graph
xmin = 0
xmax = 10
ymin = 0
ymax = 150

# For the line on the graph
y3 = m*xmin + b
y4 = m*xmax + b

# Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax],'b') # blue y axis

# Add details to the graph
ax.set_xlabel("time")
ax.set_ylabel("population")
ax.grid(True)
# ax.set_xticks(np.arange(xmin, xmax, 2))
# ax.set_yticks(np.arange(xmin, xmax, 2))

# Plot the linear function as a red line
plt.plot([xmin,xmax],[y3,y4],'r')

plt.show()

```

Solve the equation - Review exercise #26

```python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = 875 - 50*x

solution = solve(eq,x)
print("x = ",solution[0])
```
